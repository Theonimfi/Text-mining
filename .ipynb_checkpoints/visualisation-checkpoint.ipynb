```python
triples = [{'head': 'Poseidon', 'type': 'instance of', 'tail': 'sea god'},
{'head': 'Ares Anchiroe', 'type': 'father', 'tail': 'Sithon'},
{'head': 'Ares', 'type': 'sibling', 'tail': 'Zeus'}, {'head': 'Zeus', 'type': 'sibling', 'tail': 'Ares'},
{'head': 'Mary', 'type': 'sibling', 'tail': 'Diana'}, {'head': 'Diana', 'type': 'sibling', 'tail': 'Mary'},
{'head': 'Mary', 'type': 'spouse', 'tail': 'Jonathan'}, {'head': 'Jonathan', 'type': 'spouse', 'tail': 'Mary'},
{'head': 'Barack Obama', 'type': 'vice president', 'tail': 'Joe Biden'}, {'head': 'Joe Biden', 'type': 'position held', 'tail': 'vice president'},
{'head': 'Peter', 'type': 'employer', 'tail': "Mike's boss"}, {'head': "Mike's boss", 'type': 'officeholder', 'tail': 'Peter'}]
```

```python
import matplotlib.pyplot as plt
import networkx as nx

edges = [(t['head'], t['tail']) for t in triples]
G = nx.Graph()
G.add_edges_from(edges)
pos = nx.spring_layout(G, k=5, scale=0.8)
plt.figure()
nx.draw(
    G,
    pos,
    edge_color='black',
    width=1,
    linewidths=1,
    node_color='pink',
    labels={node: node for node in G.nodes()}
)
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels={(t['head'], t['tail']): t['type'] for t in triples},
    font_color='grey'
)
plt.axis('off')
plt.show()
```
